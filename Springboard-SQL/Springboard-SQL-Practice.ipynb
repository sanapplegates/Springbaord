#### Q7: Produce a list of all members who have used a tennis court. Include in your output the name of the court, and the name of the member formatted as a single column. Ensure no duplicate data, and order by the member name.

```python
query7 ='''SELECT DISTINCT m.firstname as Name,f.name as TennisCourtUsed
            FROM Members AS m
            INNER JOIN Bookings  AS b USING(memid)
            INNER JOIN Facilities  AS f USING(facid)
            WHERE  f.name LIKE 'Tennis%'
            ORDER BY Name
            '''
query7_df =pd.read_sql_query(query7,connection)
query7_df
```

#### Q8: Produce a list of bookings on the day of 2012-09-14 which will cost the member (or guest) more than $30. Remember that guests have different costs to members (the listed costs are per half-hour 'slot'), and the guest user's ID is always 0. Include in your output the name of the facility, the name of the member formatted as a single column, and the cost. Order by descending cost, and do not use any subqueries. 

```python
query8 = '''SELECT name, m.firstname||' '||m.surname as MemmberName, guestcost
            FROM Bookings
            INNER JOIN Facilities USING (facid)
            INNER JOIN Members AS m USING (memid)
            WHERE starttime LIKE '2012-09-14%'  AND guestcost > 30
            ORDER BY guestcost DESC
            '''
query8_df = pd.read_sql_query(query8,connection)
query8_df
```

#### Q9: This time, produce the same result as in Q8, but using a subquery.

```python
query9 = '''SELECT name,m.firstname||' '||m.surname as MemmberName, guestcost
            FROM Bookings
            INNER JOIN Facilities USING (facid)
            INNER JOIN Members m USING (memid)
            WHERE starttime LIKE '2012-09-14%'  AND facid IN
            (SELECT facid 
            FROM Facilities
            WHERE guestcost > 30)
            '''
query9_df =pd.read_sql_query(query9,connection)
query9_df
```

#### Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members

```python
query10 ='''SELECT Facilities.name AS FacilityName,
             SUM(
                   CASE WHEN Bookings.memid='0' 
                        THEN Facilities.guestcost * Bookings.slots 
                        ELSE Facilities.membercost * Bookings.slots 
                    END) AS TotalCost
             
            FROM Bookings
            INNER JOIN Facilities USING(facid) 
            GROUP BY FacilityName
            HAVING TotalCost<1000
            ORDER BY TotalCost DESC
            '''
query10_df =pd.read_sql_query(query10,connection)
query10_df
```

#### Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order 

```python
query11 ='''SELECT m1.surname,m1.firstname,m2.firstname||' '||m2.surname as RecommendedBy 
            FROM Members AS m1
            INNER JOIN Members AS m2
            ON m1.recommendedby=m2.memid
            ORDER BY m1.surname,m1.firstname 
           '''
query11_df = pd.read_sql_query(query11,connection)
query11_df
```

####  Q12: Find the facilities with their usage by member, but not guests

```python
query12 ='''SELECT name, COUNT(
            CASE WHEN memid > 0 THEN 1            /* memid for guests is 0*/
                else null end) AS usage
        FROM Facilities AS f
        INNER JOIN Bookings  USING(facid)
        GROUP BY name
        ORDER BY usage DESC
        '''
query12_df = pd.read_sql_query(query12,connection)
query12_df
```

#### Q13: Find the facilities usage by month, but not guests

```python
query13 ='''SELECT name, strftime('%m', starttime) AS month, COUNT(
            CASE WHEN memid > 0 THEN 1     /* memid for guests is 0 */
                ELSE NULL END) AS usage
            FROM Facilities AS f
            INNER JOIN Bookings USING (facid)
            GROUP BY name, month
            ORDER BY usage DESC
            '''
query13_df = pd.read_sql_query(query13,connection)
query13_df
```
